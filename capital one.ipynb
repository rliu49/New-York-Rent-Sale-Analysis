# Airbnb Data
Get the rent data for each zipcode in NY

```python
# read airbnb dataset
airBnb = pd.read_csv("listings.csv")
airBnb.head(3)
```

```python
airBnb.columns
```

```python
# Data Cleaning
# Keep the related columns
airBnb.isnull().sum()
```

```python
# keep the apartment that has 2 bedrooms and remove the others
airBnb_2Bed = airBnb[airBnb["bedrooms"] == 2]
airBnb_2Bed.head(3)
```

```python
# keep the relevent columns: city, zipcode, price
Bedrooms_preclean = airBnb_2Bed[['zipcode','price']]
Bedrooms_preclean.head(5)
```

```python
Bedrooms_preclean.isnull().sum()
```

```python
# read NY zipcode file and find if there is any match.
NY_zipcode = pd.read_csv('New York City Zipcode.csv')
NY_zipcode.head(3)
```

```python
# According to city to find the zipcode , and then drop the city columns.
zipcode_NA = Bedrooms_preclean[Bedrooms_preclean['zipcode'].isnull()]
zipcode_NA.head(5)
# 62 zipcodes missing out of 4894 samples.
```

```python
Airbnb_clean = Bedrooms_preclean.dropna() 
Airbnb_clean.head(3)
#4832
```

```python
Airbnb_clean.columns[1:]
```

```python
Airbnb_clean['price'] = Airbnb_clean[Airbnb_clean.columns[1:]].replace('[\$,]', '', regex=True).astype(float)
```

```python
Airbnb_clean
```

```python
Airbnb_clean = Airbnb_clean.groupby('zipcode').mean()
Airbnb_clean.head(3)
```

```python
Airbnb_clean.zipcode  = Airbnb_clean.zipcode.astype(str)
```

```python
zillowNYclean.zipcode = zillowNYclean.zipcode.astype(str)
```

```python
# merge two dataframes: airbnb_clean 186 and zillowNYclean 475 on zipcode = zipcode  
df_new = pd.merge(Airbnb_clean, zillowNYclean, how="inner", on=['zipcode'])
df_new
```

```python
# Calculate Rent to Sell Ratio by rent price/selling price at 2017-06
df_new['RS_ratio'] = df_new['price']/df_new['2017-06']
df_new
```
